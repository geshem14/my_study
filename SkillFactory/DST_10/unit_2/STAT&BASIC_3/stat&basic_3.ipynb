## Задание 3.4.1 
Аккумуляторный завод производит батарейки, срок службы которых имеет стандартное отклонение 2.4 месяца. Средний срок службы батареек в случайной выборке из 64 штук составил 12.35 месяца. Необходимо построить 90 %-ный доверительный интервал для истинного среднего срока службы производимой батарейки и ответить на вопросы заданий ниже.  
Какое критическое значение необходимо использовать? Введите число с двумя знаками после разделительной точки.

```python
sigma = 2.4  # стандартное отклонение

n = 64  # элементов в выборке
E_sampling_64 = 12.35  # матожидание выборки

alpha = 0.1  # Необходимо построить 90 %-ный доверительный интервал. alfa=1-0.9
z_0_05 = norm.ppf(1- .05)  # z критическое для alfa/2=0.05 

# проверяем допущения центральной предельной теоремы:
# Случайна ли выборка? Да.
# Достаточно ли велика выборка? да n>30

loc_left = E_sampling_64-z_0_05*sigma/(n**0.5)
loc_right = E_sampling_64+z_0_05*sigma/(n**0.5)

task = "3.4.1"
print(f"Ответ к заданию {task}")
print(f"критическое значение:= {round(z_0_05,2)}")
print(f"90 %-ный доверительный интервал для истинного среднего срока службы производимой батарейки:")
print(f"    Нижняя граница:= {round(loc_left,3)}")
print(f"    Верхняя граница:= {round(loc_right,3)}")
```

```python
def confidence_interval_norm(alpha, sigma, n, mean):
    value = -norm.ppf(alpha / 2) * sigma / (n**0.5)
    return mean - value, mean + value
```

```python
def confidence_interval_t(alpha, s, n, mean):
    value = -t.ppf(alpha / 2, n - 1) * s / (n**0.5)  # обратите внимание в курсе ошибка 
                                                     # перед t нет минуса
                                                     # в результате верхняя и нижняя граница перепутаны
    return mean - value, mean + value
```

Давайте поможем менеджеру ответить на несколько вопросов!  
Чему равна выборочная пропорция положительных комментариев? Укажите с точностью до трех знаков после разделительной точки.

```python
n = 189  # элементов в выборке
p_positive = 132/189  # 132 комментария были положительными

task = "3.8.1"
print(f"Ответ к заданию {task}")
print(f"выборочная пропорция положительных комментариев:= {round(p_positive,3)}")
```

```python
# созданим функцию для расчета доверительных интервалов пропорции
def confidence_interval_norm_proportion(alpha, p, n):
    value = -norm.ppf(alpha / 2) * ((p * (1 - p) / n)**0.5)
    return p - value, p + value
```

## 3.8.2 Задание 2
Найдите 90 %-ный доверительный интервал для истинной пропорции положительных комментариев. Ответ введите с точностью до трех знаков после разделительной точки.

```python
p = p_positive  # смотри предыдущую задачу

task = "3.8.2"
print(f"Ответ к заданию {task}")

procent = 90
alpha = 1-procent/100  
loc_left = confidence_interval_norm_proportion(alpha, p, n)[0]
loc_right = confidence_interval_norm_proportion(alpha, p, n)[1]
print(f"{procent} %-ный доверительный интервал:")
print(f"    Нижняя граница:= {round(loc_left,4)}")
print(f"    Верхняя граница:= {round(loc_right,3)}")
```

```python
# содадим функцию для z критического от процента
def z_critical(d_procent):
    d_alpha = 1-d_procent/100
    return -norm.ppf(d_alpha / 2)
```

```python
# содадим функцию для z критического для двустороннего t-теста
def z_critical_t_test(d_procent, d_degreeOfFreedom):
    d_alpha = 1-d_procent/100
    return -t.ppf(d_alpha/2, d_degreeOfFreedom)
```
