## Задание 4.7.1  
Условие:  
Производитель аккумуляторных батарей заявляет, что средний срок службы батарейки АА составляет 14 месяцев. Вы подозреваете, что на самом деле это число преувеличено.  
Чтобы проверить утверждение производителя, вы взяли случайную выборку из 64 батареек и посчитали выборочное среднее — 13.5 месяцев. Известно, что срок службы батарейки имеет стандартное отклонение 2 месяца. Для тестирования гипотезы используйте уровень значимости (alpha =0.05).  
Обратите внимание, что как постановка альтернативной гипотезы, так и выбор уровня значимости alpha может повлиять на вывод, который вы делаете в результате тестирования гипотезы.

1. Первый шаг при тестировании – постановка нулевой и альтернативной гипотез. Нулевая гипотеза: μ=14. Какую альтернативную гипотезу нужно использовать в данном случае?  

Ответ: μ<14

```python
# создадим функцию для z z-статистики
def z_statistic(E, mu_0, sigma, n):
    return (E-mu_0)/(sigma/(n**0.5))
```

## Задание 4.7.2
Чему равна z-статистика?

```python
n = 64  # элементов в выборке
E_sampling_64 = 13.5  # матожидание выборки

sigma = 2  # стандартное отклонение 
mu_0 = 14

z_stat = z_statistic(E_sampling_64, mu_0, sigma, n)

# проверяем допущения центральной предельной теоремы:
# Случайна ли выборка? Да.
# Достаточно ли велика выборка? да n>30  

task = "4.7.2"
print(f"Ответ к заданию {task}")
print(f"z-статистика равна:= {round(z_stat,0)}")
```

## Задание 4.7.3
Чему равно значение p-value (ответ округлите до 3 знаков после разделительной точки)?

```python
p_value = norm.cdf(z_stat)

task = "4.7.3"
print(f"Ответ к заданию {task}")
print(f"значение p-value равно:= {round(p_value,3)}")
```

## Задание 4.10.1  
Условие:  
Представитель профсоюза утверждает, что 75% сотрудников готовы присоединиться к забастовке, если их требования не будут удовлетворены. Сотрудник компании, ведущий переговоры с профсоюзом, считает, что эта пропорция сильно преувеличена. Он решает проверить утверждение представителя профсоюза, используя уровень значимости 5%.  
Таким образом, ему надо протестировать следующий набор гипотез: 

H~0~ : p = 0.75, H~A~ : p < 0.75

В выборке из 125 сотрудников 87 ответили, что будут готовы поддержать забастовку в случае неудовлетворения требования профсоюза.  
Чему равна z-статистика для данного теста? Ответ округлите до 3 знаков после разделительной точки.

```python
p_0 = 0.75  # Мы хотим проверить, не снизилась ли уже наша пропорция ниже 0.75

# cобираем наблюдения
p_ = 87/125
n = 125

# cтроим наши ожидания. 
# выборка случайная
# достаточно большая: n*p_==87 > 5
n*(1-p_)
```

```python
# n*(1-p_) > 5

# Предполагаем, что нулевая гипотеза верна и заменяем p на p_0:
z_stat_prop = (p_ - p_0)/(((p_0*(1-p_0))/n)**0.5)

task = "4.10.1"
print(f"Ответ к заданию {task}")
print(f"z-статистика для данного теста равна:= {round(z_stat_prop,3)}")
```
