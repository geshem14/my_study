### Задание 5.5.1  
Среди людей заходящих в магазин, 70% ничего не покупает. Остальные с равной вероятностью покупают одну-три бутылочки воды.  
Найдите вероятность купить две бутылочки, если известно, что их было куплено чётное число:  
***Решение:***  
$P(A|B) = \frac{P(A \cap B)}{P(B)} = P(\{2\}|\{0,2\})$  
$P(A \cap B)= P(\{2\}) = 1/10$  
$P(B)= P(\{0,2\})= 8/10$  
$P(\{2\}|\{0,2\}) = \frac{1/10}{8/10} = 1/8$  

***Ответ:***  1/8 = 0.125

### Задание 5.5.2
Среди тех, кому показали рекламный баннер, 20 % оказались мужчинами. Мужчины переходят по ссылке баннера с вероятностью 5%, а женщины — с вероятностью 2.5%  
1) С какой вероятностью пользователь, которому показали рекламу, окажется женщиной, кликнувшей по ссылке?
***Решение:***  
если M - событие - показали рекламный баннер чеовеку и он оказался мужчиной, тогда P(M)=0.2  
тогда $\overline{M}$  - событие - показали рекламный баннер человеку и он оказался женщиной:  
$P(\overline{M}) = 1 - 0.2 = 0.8 = P(W)$  
Теперь нам нужно пересечь множество всех этих женщин с теми кто кликнул  
$P(W \cap C)= P(C|W)*P(W) = 0.025 * 0.8 = 0.02$  

***Ответ:***  0.02

### 5.10. Независимость событий
---
В приложениях теории вероятностей к анализу данных очень важным является понятие независимости. Сейчас мы говорим о вероятности событий, поэтому и независимость рассмотрим тоже для событий.  
Определение:  События А и В называются независимыми, если вероятность их пересечения равна произведению вероятностей  

$P(A\cap B)=P(A)\cdot P(B)$  

Главное свойство независимых событий: Вероятность А не зависит от наличия условия. То есть при наступлении события В событие А происходит так же часто, как и без него  

$P(A|B)=P(A)$

```python
# функция которая выдает индексы массива по условию
def index_vec(d_v, d_bool_str,d_cond):
    if d_bool_str == '==':
        full_cond = (d_v == d_cond)
    elif d_bool_str == '>=':
        full_cond = (d_v >= d_cond)
    elif d_bool_str == '>':
        full_cond = (d_v > d_cond)
    elif d_bool_str == '<=':
        full_cond = (d_v <= d_cond)
    elif d_bool_str == '<=':
        full_cond = (d_v >= d_cond)
    
    temp = np.where(full_cond)
    
    return temp
```

```python
ind_X = index_vec(X_values, '==', 0)
ind_Y = index_vec(Y_values, '==', 1)
temp = probability_X_Y[ind_X,ind_Y].sum()

task = "5.17.1"
print(f"Ответ к заданию {task} пункт 3.1")
print(f"P(X=0, Y=1):= {temp}")
```

### Задание 5.17.3
Совместное распределение случайных величин  задано в таблице (см. код)  
1. Найдите cov(X,Y), округлив до двух знаков после точки  
2. Найдите corr(X,Y), округлив до трёх знаков после точки

```python
X_values = np.array([-1, 0, 1])
Y_values = np.array([0, 1])

probability_X_Y = np.array([
                [.1, .1],
                [.3, .1],
                [.2, .2]])
X, Y = margins(probability_X_Y)

E_XY = X_values @ probability_X_Y @ Y_values
E_X = X_values @ X
E_Y = Y_values @ Y.T
temp = E_XY-E_X*E_Y

sigma_X = X_values*X_values @ X - E_X**2
sigma_X = sigma_X**0.5
sigma_Y = Y_values*Y_values @ Y.T - E_Y**2
sigma_Y = sigma_Y**0.5

task = "5.17.3"
print(f"Ответ к заданию {task} пункт 4")
print(f"cov(X,Y):= {round(temp[0],2)}")
print(f"corr(X,Y):= {round(temp[0]/(sigma_X[0]*sigma_Y[0]),3)}")
```
